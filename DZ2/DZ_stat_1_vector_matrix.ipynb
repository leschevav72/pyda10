## Задание 1

Изобразите с помощью matplotlib. Изобразите точку x + 2y + 3z. Найдите угол между векторами x, y и x, z. 

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
xy=sum([a*b for a,b in zip(x,y)])
x_len=sum([a**2 for a in x])**0.5
y_len=sum([a**2 for a in y])**0.5
z_len=sum([a**2 for a in z])**0.5
XYalpha=math.acos(xy/(x_len*y_len))
xz=sum([a*b for a,b in zip(x,z)])
XZalpha=math.acos(xz/(x_len*z_len))
print('угол между векторами x,y =',XYalpha,'угол между векторами x,z =', XZalpha)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

x = np.array([1])
y = np.array([2])
z = np.array([3])

ax.plot([0,x[0]], [0,y[0]], [0,z[0]])


ax.scatter(x[0], y[0], z[0], c='b')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```

```python
fig, axs=plt.subplots(1,1,figsize=(17,8))
axs.plot([0,x[0]], [0,y[0]], label='в плоскости XY')
axs.plot([0,y[0]], [0,z[0]],label='в плоскости YZ')
axs.plot([0,x[0]], [0,z[0]],label='в плоскости XZ')

axs.legend(loc='best')
plt.show()
```
