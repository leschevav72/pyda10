```python
a={}
#i=0
with open('purchase_log.txt', 'r') as f:
    for line in f:
        line = line.strip()
        dict_ = json.loads(line)
        a[dict_['user_id']]=dict_['category']
        #i += 1
        #if i > 5:
            #break
#print(a)
        
```

```python
output=open('funnel.csv', 'w')
with open('visit_log.csv', 'r') as inp:
    for entry in inp.readlines():
        uid, source=entry.strip().split(',')
        if uid in a.keys():
            output.write('{},{},{}\n'.format(uid, a[uid], source))
output.close()
```
